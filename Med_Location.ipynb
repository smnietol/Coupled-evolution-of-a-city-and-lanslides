```python
# Load world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Set up the plot
fig, ax = plt.subplots(subplot_kw=dict(projection=ccrs.Orthographic(central_longitude=-74, central_latitude=15)))
ax.spines['geo'].set_edgecolor('gray')

# Plot the world
world.plot(ax=ax, transform=ccrs.PlateCarree(), color='silver', edgecolor='lightgray', linewidth=0.5)

# Highlight Colombia
colombia = world[world.name == 'Colombia']
colombia.plot(ax=ax, transform=ccrs.PlateCarree(), color='lightgray', edgecolor='darkorange')
ax.plot(-75.5812, 6.2442, marker='o', color='orange', markersize=2,
        alpha=0.7, transform=ccrs.Geodetic(), label='Colombia')
#Plot Medelln
ax.plot(-75.5812, 6.2442, marker='o', color='red', markersize=2,
        alpha=0.7, transform=ccrs.Geodetic(), label='Medelln City')

# Add gridlines
ax.gridlines(draw_labels=False, color='gray', alpha=0.5, linestyle='--', linewidth=0.1)

legend_handles = [
    Line2D([0], [0], marker='o', color='w', markerfacecolor='orange', markersize=8, label='Colombia'),
    Line2D([0], [0], marker='o', color='w', markerfacecolor='red', markersize=8, label='Medelln City')]

#plt.legend(handles=legend_handles, bbox_to_anchor=(1.3, 1))

plt.savefig("Med_location.pdf", dpi=300, bbox_inches='tight')
plt.show()
```
